```python
import graphlab
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = graphlab.SFrame.read_csv("1m.csv",
                                column_type_hints={"UserID": int, 
                                                   "MovieTitle": str, 
                                                   "Ratings": int})
```

```python
data = data.select_columns(["UserID", "MovieTitle", "Ratings"])
```

```python
data.head()
```

# Making recommendations to new users

## Rate movies

### Action lover

```python
action_movies = graphlab.SArray(["GoldenEye (1995)",
                                 "Braveheart (1995)",
                                 "Terminator 2: Judgment Day (1991)",
                                 "Die Hard (1988)",
                                 "Top Gun (1986)",
                                 "Indiana Jones and the Last Crusade (1989)",
                                 "Crow, The (1994)",
                                 "Cop Land (1997)",
                                 "Mad Love (1995)",
                                 "Sabrina (1995)",
                                 "While You Were Sleeping (1995)",
                                 "Sleepless in Seattle (1993)",
                                 "Dirty Dancing (1987)",
                                 "When Harry Met Sally... (1989)",
                                 "Jerry Maguire (1996)",
                                 "Pretty Woman (1990)"],
                                str)
action_ratings = graphlab.SArray([5, 4, 5, 5, 3, 4, 4, 5, 1, 2, 1, 2, 1, 1, 2, 1], int)
action_userid = graphlab.SArray([9000] * action_movies.size(), int)

action_lover = graphlab.SFrame()
action_lover.add_column(action_userid, "UserID")
action_lover.add_column(action_movies, "MovieTitle")
action_lover.add_column(action_ratings, "Ratings")
```

### Romance lover

```python
romance_movies = graphlab.SArray(["Mad Love (1995)",
                                  "Sabrina (1995)",
                                  "While You Were Sleeping (1995)",
                                  "Sleepless in Seattle (1993)",
                                  "Dirty Dancing (1987)",
                                  "When Harry Met Sally... (1989)",
                                  "Jerry Maguire (1996)",
                                  "Pretty Woman (1990)",
                                  "GoldenEye (1995)",
                                  "Braveheart (1995)",
                                  "Terminator 2: Judgment Day (1991)",
                                  "Die Hard (1988)",
                                  "Top Gun (1986)",
                                  "Indiana Jones and the Last Crusade (1989)",
                                  "Crow, The (1994)",
                                  "Cop Land (1997)"],
                                 str)
romance_ratings = graphlab.SArray([5, 4, 5, 3, 4, 5, 5, 4, 2, 2, 1, 1, 2, 2, 1, 1], int)
romance_userid = graphlab.SArray([9001] * romance_movies.size(), int)

romance_lover = graphlab.SFrame()
romance_lover.add_column(romance_userid, "UserID")
romance_lover.add_column(romance_movies, "MovieTitle")
romance_lover.add_column(romance_ratings, "Ratings")
```

## Recommend movies
We need to update the model with the ratings of the action and romance lovers.

```python
df = data.to_dataframe()
df = df.append(action_lover.to_dataframe())
df = df.append(romance_lover.to_dataframe())
new_data = graphlab.SFrame(data=df)
```

```python
new_data.tail(32)
```

```python
optimal_model = graphlab.recommender.matrix_factorization.create(new_data, "UserID", "MovieTitle", "Ratings", D=150, reg=0.1)
```

```python
optimal_model.training_stats()
```

```python
optimal_stats = optimal_model.evaluate(new_data)
```

### Action lover
The top ten recommendations for the action lover are the following:

```python
action_recommendations = optimal_model.recommend(new_data, users=[9000], k=10)
action_recommendations.head(10)
```

### Romace lover
The top ten recommendations for the romance lover are the following:

```python
romance_recommendations = optimal_model.recommend(new_data, users=[9001], k=10)
romance_recommendations.head(10)
```
